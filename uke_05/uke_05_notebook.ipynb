```python
# list for timepoints
t = [0, 30, 60, 90, 120, 150, 180, 210, 240]
# list for number of bacteria at 39 degrees
E = [10000, 10000, 10000, 25587, 76327, 212715, 619511, 1940838, 4240760]

#  data for number of bacteria at 29 degrees
t2 = [0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360, 390, 420]
E2 = [10000, 10000, 10000, 10000, 19121, 28904, 49730, 75381, 136761, 201330, 323768, 652591, 955065, 1698947, 3310323]
```

```python
pyplot.plot(t, E, label = "Growth at 39 C")
pyplot.plot(t2, E2, label = "Growth at 29 C")
pyplot.title("Bacterial growth at different temperatures")
pyplot.xlabel("Time (minutes)")
pyplot.ylabel("Population size, E")
pyplot.legend()
pyplot.show()
```
