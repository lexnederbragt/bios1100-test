## Human population size

The table below shows the population size in Norway between 1952 and 2007 with
five-year intervals:

| Year | Population |
|------|---------|
| 1952 | 3327728 |
| 1957 | 3491938 |
| 1962 | 3638919 |
| 1967 | 3786019 |
| 1972 | 3933004 |
| 1977 | 4043205 |
| 1982 | 4114787 |
| 1987 | 4186147 |
| 1992 | 4286357 |
| 1997 | 4405672 |
| 2002 | 4535591 |
| 2007 | 4627926 |

**Exercise 1**
Generate a line plot of the population over time, with figure title and axes
lables.

```python
# Solution
%matplotlib inline
from matplotlib import pyplot

Y = [1952, 1957, 1962, 1967, 1972, 1977, 1982, 1987, 1992, 1997, 2002, 2007]
P = [3327728, 3491938, 3638919, 3786019, 3933004, 4043205, 4114787, 4186147, 4286357, 4405672, 4535591, 4627926]

pyplot.plot(Y, P)
pyplot.title("Population size in Norway 1952-2007")
pyplot.xlabel("Year")
pyplot.ylabel("Population size")
pyplot.show()
```

**Exercise 2**
Generate a new plot of showing population sizes in both countries over time,
also include a legend.

```python
# Solution
# NOTE it is not necessary to redo the follwoing two lines
# but they are included for completeness
%matplotlib inline
from matplotlib import pyplot

Y = [1952, 1957, 1962, 1967, 1972, 1977, 1982, 1987, 1992, 1997, 2002, 2007]
PNorway = [3327728, 3491938, 3638919, 3786019, 3933004, 4043205, 4114787, 4186147, 4286357, 4405672, 4535591, 4627926]
PDenmark = [4334000, 4487831, 4646899, 4838800, 4991596, 5088419, 5117810, 5127024, 5171393, 5283663, 5374693, 5468120 ]

pyplot.plot(Y, PNorway, label = "Norway")
pyplot.plot(Y, PDenmark, label = "Denmark")
pyplot.title("Population size 1952-2007")
pyplot.xlabel("Year")
pyplot.ylabel("Population size")
pyplot.legend()
pyplot.show()
```

**Exercise 4**
Using the help text of `pyplot.legend()`, find out how to improve the figure by
placing the legend in a different part of the plot.

```python
# Solution
# NOTE it is not necessary to redo the follwoing two lines
# but they are included for completeness
%matplotlib inline
from matplotlib import pyplot

Y = [1952, 1957, 1962, 1967, 1972, 1977, 1982, 1987, 1992, 1997, 2002, 2007]
PNorway = [3327728, 3491938, 3638919, 3786019, 3933004, 4043205, 4114787, 4186147, 4286357, 4405672, 4535591, 4627926]
PDenmark = [4334000, 4487831, 4646899, 4838800, 4991596, 5088419, 5117810, 5127024, 5171393, 5283663, 5374693, 5468120 ]

pyplot.plot(Y, PNorway, label = "Norway")
pyplot.plot(Y, PDenmark, label = "Denmark")
pyplot.title("Population size 1952-2007")
pyplot.xlabel("Year")
pyplot.ylabel("Population size")
pyplot.legend(loc='upper left')
pyplot.show()
```
